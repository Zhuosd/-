```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import cv2
import os
```

```python
# 去除背景，获取单独的车牌图片
def find_card(I):
    # 识别出车牌区域并返回该区域的图像
    [y, x, z] = I.shape
    # y取值范围分析
    Blue_y = np.zeros((y, 1))
    for i in range(y):
        for j in range(x):
            # 蓝色rgb范围
            temp = I[i, j, :]
            if (I[i, j, 2] <= 30) and (I[i, j, 0] >= 119):
                Blue_y[i][0] += 1
    # 获取B维度的图像数据
    MaxY = np.argmax(Blue_y)
    PY1 = MaxY
    while (Blue_y[PY1, 0] >= 5) and (PY1 > 0):
        PY1 -= 1
    PY2 = MaxY
    while (Blue_y[PY2, 0] >= 5) and (PY2 < y - 1):
        PY2 += 1
    # x取值
    Blue_x = np.zeros((1, x))
    for i in range(x):
        for j in range(PY1, PY2):
            if (I[j, i, 2] <= 30) and (I[j, i, 0] >= 119):
                Blue_x[0][i] += 1
    PX1 = 0
    while (Blue_x[0, PX1] < 3) and (PX1 < x - 1):
        PX1 += 1
    PX2 = x - 1
    while (Blue_x[0, PX2] < 3) and (PX2 > PX1):
        PX2 -= 1
    # 对车牌区域的修正
    PX1 -= 2
    PX2 += 2
    
    return I[PY1:PY2, PX1 - 2: PX2, :] # 获取对应的车牌数据
```

```python
# 维度切分
def divide(I):
    [y, x, z] = I.shape # 获取三个不同的维度
    White_x = np.zeros((x, 1)) # 设置x轴的维度信息为1
    for i in range(x): # 进行维度的遍历
        for j in range(y):
            if I[j, i, 1] > 176:
                White_x[i][0] += 1
    return White_x # 返回x轴的值
```

```python
# 针对每个字符进行切分，保存为独立的每个文件
def divide_each_character(I):
    [y, x, z] = I.shape  # 获取三个维度数据
    White_x = np.zeros((x, 1)) 
    for i in range(x): # 进行遍历，进行数据的读取
        for j in range(y):
            if I[j, i, 1] > 176:
                White_x[i][0] += 1
    res = []
    length = 0
    for i in range(White_x.shape[0]): # 进行x轴数据的读取
        # 使用超参数经验分割
        t = I.shape[1] / 297
        num = White_x[i]
        if num > 8:
            length += 1
        elif length > 20 * t:
            res.append([i - length - 2, i + 2])
            length = 0
        else:
            length = 0
    return res
```

```python
if __name__ == '__main__':
    # 读取文件
    I = cv2.imread('040625-90_88-170&439_574&562-573&560_162&558_172&442_583&444-0_0_33_29_13_30_31-85-150.jpg')
    Plate = find_card(I) # 去除背景，获取单独的车牌图片
    # 显示图片
    # White_x = divide(Plate)
    plt.imshow(Plate)
    plt.show()
    
    # plt.plot(np.arange(Plate.shape[1]), White_x)
    res = divide_each_character(Plate) # 针对每个字符进行切分，保存为独立的每个文件
    # 每个文件只能保存一个，如果换一张图片，就更换另外一个文件夹
    plate_save_path = './1/'
    # 进行文字、数字、字母的切分工作
    for t in range(len(res)):
        plt.subplot(1, 7, t + 1) # 子图
        temp = res[t] # 根据t的顺序，进行保存
        save_img = cv2.cvtColor(Plate[:, temp[0]:temp[1], :],cv2.COLOR_BGR2GRAY) # 转化为二值图进行保存
        ma = max(save_img.shape[0], save_img.shape[1]) # 图像维度最大一维
        mi = min(save_img.shape[0], save_img.shape[1]) # 图像维度最小一维
        ans = np.zeros(shape=(ma, ma, 3),dtype=np.uint8) # 生成三个维度且都为0的数组
        start =int(ma/2-mi/2)
        # 进行图像的划分
        for i in range(mi):
            for j in range(ma):
                if save_img[j,i] > 125:
                    for k in range(3):
                        ans[j,start+i,k]=255
        ans=cv2.merge([ans[:,:,0],ans[:,:,1],ans[:,:,2]]) # 获取三个维度的数据信息
        ans=cv2.resize(ans,(25,25)) # 调整大小为25*25
        dir_name=plate_save_path+str(t) # 保存路径
        os.mkdir(dir_name)
        cv2.imwrite(dir_name+'/'+str(t)+'.jpg',ans) # 保存图像
        print("保存路径为",dir_name+'/'+str(t)+'.jpg')  # 打印所切割图片的位置
        plt.imshow(ans)
    # 显示最终的切割图片
    plt.show()
```
